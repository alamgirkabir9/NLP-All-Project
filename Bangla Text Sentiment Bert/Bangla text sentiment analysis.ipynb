```python
# Text preprocessing function
def preprocess_text(text):
    if pd.isnull(text) or not isinstance(text, str):
        return None
    # Replace emojis with corresponding text
    for emoji, meaning in emoji_dict.items():
        text = text.replace(emoji, meaning)
    # Remove stopwords
    words = text.split()
    words = [word for word in words if word not in stopwords]
    # Remove any extra spaces
    text = " ".join(words)
    return text.strip() if text else None
```

```python
import re

def clean_text(text):
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^ঀ-৺]', ' ', text)  # Keep only Bangla characters, replace others with space
    words = text.split()                 # Tokenize into words
    text = " ".join(words)              # Join words with single space
    text = text.strip()                 # Remove leading/trailing whitespace
    return text
```

```python
import torch
from transformers import AutoTokenizer, BertForSequenceClassification

# Load the fine-tuned model and tokenizer
model = BertForSequenceClassification.from_pretrained("/content/drive/MyDrive/NLP/bangla_bert_sentiment_model3")
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/NLP/bangla_bert_sentiment_model3")

# Define a dictionary to map labels back to sentiment classes
label_mapping = {0: "happy", 1: "sad", 2: "angry"}

# Define a function to preprocess the input text
def preprocess_text(text):
    stopwords = ['আমি', 'আপনি', 'সে', 'তুমি']
    emoji_dict = {
        "😊": "হাসি", "😢": "কান্না", "😠": "রাগ", "😔": "মন খারাপ", "👍": "দারুণ",
        "😎": "ঠান্ডা", "😭": "অশ্রু", "😁": "মুচকি হাসি", "😅": "হালকা হাসি", "😍": "ভালবাসা",
        "😒": "অসন্তুষ্ট", "😞": "হতাশা", "😡": "রাগান্বিত", "😃": "খুশি", "😉": "চোখ মারা",
        "😋": "স্বাদ আস্বাদন", "😐": "নির্বিকার", "😤": "অসন্তুষ্ট", "😴": "ঘুম", "😜": "মজার",
        "😩": "ক্লান্ত", "😯": "আশ্চর্য", "😆": "হাসি", "😷": "মাস্ক", "🙄": "গুর্গুরানি",
        "😳": "বিস্মিত", "😬": "চিন্তিত", "😚": "চুম্বন", "😰": "উদ্বিগ্ন", "🤗": "আলিঙ্গন",
        "🤔": "চিন্তাশীল", "🤐": "মুখ বন্ধ", "😇": "পুণ্যবান"
    }

    # Replace emojis with corresponding text
    for emoji, meaning in emoji_dict.items():
        text = text.replace(emoji, meaning)

    # Remove stopwords
    words = text.split()
    words = [word for word in words if word not in stopwords]
    text = " ".join(words)
    return text.strip()

# Define a function to predict sentiment
def predict_sentiment(text):
    # Preprocess the text
    text = preprocess_text(text)
    text = clean_text(text)
    # Tokenize the text
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)

    # Predict sentiment
    with torch.no_grad():
        model.eval()
        outputs = model(**inputs)
        logits = outputs.logits
        prediction = torch.argmax(logits, dim=1).item()

    # Return the predicted sentiment
    return label_mapping.get(prediction, "Unknown")
```

```python
# Example usage
input_text = "আমি বারবার বলার পরও কেউ আমাকে গুরুত্ব দিচ্ছে না! কেন মানুষ প্রতিশ্রুতি ভেঙে দেয়? এত কষ্ট করে কিছু করলে যদি তার মূল্যই না থাকে, তাহলে আর করার মানে কী? অন্যায়কে মেনে নেওয়া আমার স্বভাবে নেই, তাই আমি এই ব্যবস্থার বিরুদ্ধে লড়াই করবো!"
sentiment = predict_sentiment(input_text)
print(f"Sentiment: {sentiment}")
```
